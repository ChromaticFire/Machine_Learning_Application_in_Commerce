```python
tumor_df = pd.read_excel('.xlsx')
tumor_df.head()
```

```python
tumor_X = tumor_df.drop(columns='') # 0; 1
tumor_y = tumor_df['']
X_train, X_test, y_train, y_test = train_test_split(\
tumor_X, tumor_y, test_size=0.2, random_state=3)

nb_clf = GaussianNB()
nb_clf.fit(X_train,y_train)

y_pred = nb_clf.predict(X_test)

new_df = pd.DataFrame()  # DataFrame
new_df[''] = list(y_pred)
new_df[''] = list(y_test)

new_df.head()
```

```python
accuracy = accuracy_score(y_pred, y_test)
print(accuracy)
```
